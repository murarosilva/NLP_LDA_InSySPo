```python
import os
from RISparser import readris
import gensim
```

```python
from gensim.parsing.preprocessing import STOPWORDS
```

```python
stopwords = set(STOPWORDS)
my_stopwords = {'approach','paper','article','model','viewpoint','conclusion','discussion','method','introduction','section','framework'}
stopwords = set.union(stopwords,my_stopwords)
stopwords = frozenset(stopwords)
```

```python
data_folder = 'D:/Pesquisa/NLP_LDA_InSySPo/data/'
files = os.listdir(data_folder)
```

```python
f = data_folder + files[10]
fid = open(f,'r',encoding="utf8")
data = readris(fid)
```

```python
texts = []
for i,entry in enumerate(data):
    
    try:
        title = entry['title']
        abstract = entry['abstract']
        text = title + ' ' + abstract
        text = gensim.utils.simple_preprocess(text)
        text = [word for word in text if word not in stopwords]
        texts += [text]
        print(i,len(text))
    except:
        pass
```
